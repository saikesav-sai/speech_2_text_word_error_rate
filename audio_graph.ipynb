```python
import matplotlib.pyplot as plt
from scipy.io import wavfile
import numpy as np
import os

def graph(file_path,wavfiles):
    

    # Read the WAV file
    sample_rate, data = wavfile.read(file_path)

    # Create a time array in seconds
    time = np.linspace(0, len(data) / sample_rate, num=len(data))

    # Plot the audio signal
    plt.figure(figsize=(4.5, 2))
    plt.plot(time, data, linewidth=0.5, alpha=0.7, color='green')
    plt.xlabel('Time [s]')
    # plt.ylabel('Amplitude')
    plt.title(wavfiles)
    plt.show()

# Path to the WAV file
file_path_folder='audio/denosied/'

wavfiles=os.listdir(file_path_folder)

for file in wavfiles:
    file_path = os.path.join(file_path_folder, file)
    print(file_path)
    graph(file_path,file)

   

```
